```python
import pandas as pd

original_df = pd.read_csv('2022-08-03-ss.cleaned.csv')

# Remove sequences with length greater than 400
df = original_df[original_df['seq'].str.len() <= 400]

# Remove sequences with the "*" token
df = df[~df['seq'].str.contains('*')]

df
```

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test sets
groups = df.groupby('pdb_id').groups.values()
groups = [df.iloc[list(indices)] for indices in groups]
train_groups, test_groups = train_test_split(groups, test_size=0.3, random_state=42)
train_df = pd.concat(train_groups)
test_df = pd.concat(test_groups)

train_df.to_csv('less_than_400_train_data.csv', index=False)
test_df.to_csv('less_than_400_test_data.csv', index=False)
```
